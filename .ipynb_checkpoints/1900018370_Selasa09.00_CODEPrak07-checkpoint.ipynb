```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X,y = make_blobs(random_state=0, n_samples=1000)
```

```python
X
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X[:,0],X[:,1],'ro')
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_,marker='o')
plt.show()
```

```python
kmeans = KMeans(n_clusters=10)
kmeans.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_,marker='o')
plt.show()
```

```python
kmeans = KMeans (algorithm='auto',copy_x=True, init='k-means++', max_iter=50, n_clusters=10,
                n_init=10, n_jobs=1, precompute_distances='auto',random_state=None, tol=0.0001,
                verbose=0)
kmeans.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_,marker='o')
plt.show()
```

```python
kmeans = KMeans (algorithm='auto',copy_x=True, init='k-means++', max_iter=100, n_clusters=10,
                n_init=10, n_jobs=1, precompute_distances='auto',random_state=None, tol=0.0001,
                 verbose=0)
kmeans.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_,marker='o')
plt.show()
```

```python
kmeans = KMeans (algorithm='auto',copy_x=True, init='k-means++', max_iter=200, n_clusters=10,
                n_init=10, n_jobs=1, precompute_distances='auto',random_state=None, tol=0.0001,
                 verbose=0)
kmeans.fit(X)
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_,marker='o')
plt.show()
```

```python
kmeans = KMeans (algorithm='auto',copy_x=True, init='k-means++', max_iter=300, n_clusters=10,
                n_init=10, n_jobs=1, precompute_distances='auto',random_state=None, tol=0.0001,
                 verbose=0)
kmeans.fit(X)
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_,marker='o')
plt.show()
```
